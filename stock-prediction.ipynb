# Data Preparation

```python
X = data.Open.values
Y = data.Close.values
X = X.reshape(-1,1)
Y = Y.reshape(-1,1)
x_train = X[:200000]
y_train = Y[:200000]
x_test = X[200000:250000]
y_test = Y[200000:250000]
x_val = X[250000:]
y_val = Y[250000:]
```

### Data Normalization

```python
x_train = np.log(x_train)
y_train = np.log(y_train)
x_test = np.log(x_test)
y_test = np.log(y_test)
x_val = np.log(x_val)
y_val = np.log(y_val)
```

# Modelling and Evaluation
# (Ensemble Model with Linear and AdaBoost Regressors)

```python
%%time
model1 = LinearRegression(fit_intercept= True)
model2 = AdaBoostRegressor(random_state=123)
model3 = xg.XGBRegressor(random_state=123)
model = VotingRegressor(estimators=[('lr', model1), ('abr', model2), ('xgr', model3)], verbose=True)
model.fit(x_train,y_train)
test_pred = model.predict(x_test)
val_pred = model.predict(x_val)
RMSE_test = mean_squared_error(y_test, test_pred, squared = False)
RMSE_val = mean_squared_error(y_val, val_pred, squared = False)
print("R2 score for training data is {:.2f}".format(model.score(x_train, y_train)))
print("R2 score for test data is {:.2f}".format(model.score(x_test, y_test)))
print("R2 score for validation data is {:.2f}".format(model.score(x_val, y_val)))
print("============================================================================")
print("Root mean squared error for test data is {:.2f}".format(RMSE_test))
print("Root mean squared error for validation data is {:.2f}".format(RMSE_val))
```

```python
X = data.Open.values
Y = data.Close.values
X = X.reshape(-1,1,1)
Y = Y.reshape(-1,1,1)
x_train = X[:200000]
y_train = Y[:200000]
x_test = X[200000:]
y_test = Y[200000:]
x_train = np.log(x_train)
y_train = np.log(y_train)
x_test = np.log(x_test)
y_test = np.log(y_test)
```

```python
lstm = Sequential()

lstm.add(LSTM(units=200,return_sequences=True,input_shape=(x_train.shape[1], 1)))
lstm.add(Dropout(0.3))

lstm.add(LSTM(units=200,return_sequences=True))
lstm.add(Dropout(0.3))

lstm.add(LSTM(units=200,return_sequences=True))
lstm.add(Dropout(0.3))

lstm.add(LSTM(units=200))
lstm.add(Dropout(0.2))

lstm.add(Dense(units=1))

lstm.compile(optimizer='SGD',loss='mean_squared_error')

lst_model = lstm.fit(x_train,y_train,epochs=75,batch_size=2000,validation_split=0.1)
```

```python
plt.plot(lst_model.history['loss'])
plt.plot(lst_model.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```
